```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv("E:\Sumit\LSTM\LSTM_3_Years.csv", parse_dates=True)

# Feature list
features = ['Precipitation', 'Temperature', 'SurfacePressure', 'RelativeHumidity',
            'ZTD', 'Reflectivity', 'UComponentofWind']

# Optional: apply smoothing
df_rolling = df[features].rolling(window=7, center=True).mean()

# Plot settings
plt.figure(figsize=(20, 22))
colors = {
    'Precipitation': 'red',
    'Temperature': 'blue',
    'SurfacePressure': 'purple',
    'RelativeHumidity': 'orange',
    'ZTD': 'green',
    'Reflectivity': 'brown',
    'UComponentofWind': 'black'
}

for i, feature in enumerate(features):
    plt.subplot(len(features), 1, i + 1)
    plt.plot(df_rolling[feature], label=feature, color=colors.get(feature, 'teal'))
    plt.title(f"{feature} Time Series (Smoothed)", fontsize=20) #, fontweight='bold')
    plt.ylabel(feature, fontsize=18)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.grid(True)
    plt.legend(fontsize=16, loc='upper right')

# Label the last subplot
plt.xlabel("Time Step", fontsize=18)
plt.tight_layout()
plt.savefig("timeseries_all_features_enhanced.png", dpi=600, bbox_inches='tight', transparent=True)
plt.show()
```
